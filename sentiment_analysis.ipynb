#Loading and preprocessing the data

```python
data = pd.read_csv('dataset.csv')
data
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  words = word_tokenize(text)
  words = [lemmatizer.lemmatize(word) for word in words if word.isalnum()]
  words = [word.lower() for word in words if word.lower() not in stop_words]
  return ' '.join(words)
```

```python
data['preprocessed_text'] = data['review'].apply(preprocess_text)
data
```

# Aspect Extraction

```python
nlp = spacy.load('en_core_web_sm')

def extract_aspects(text):
  doc = nlp(text)
  aspects = [chunk.text for chunk in doc.noun_chunks]
  return aspects

data['aspects'] = data['preprocessed_text'].apply(extract_aspects)
data
```

# Opinion Identification

```python
from numpy import printoptions
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

def identify_opinions(aspects):
  opinions={}

  for aspect in aspects:
    sentiment = sid.polarity_scores(aspect)['compound']

    if(sentiment>0.2):
      opinions[aspect] = 'positive'
    elif(sentiment < -0.2):
      opinions[aspect] = 'negative'
    else:
      opinions[aspect] = 'neutral'
  return opinions


for index, row in data.iterrows():
    opinions = row['opinions']
    for aspect, opinion in opinions.items():
        print(f"{opinion}")
    print()


```

# Text Summarization

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

def generate_summary(text):
  parser = PlaintextParser.from_string(text , Tokenizer("english"))
  summarizer = LsaSummarizer()
  summary = summarizer(parser.document , sentences_count=1)
  return ' '.join([str(sentence) for sentence in summary])

data['summary'] = data['review'].apply(generate_summary)
data['summary']
```

# Opinion Summary

```python
from collections import Counter

positive_counter = Counter()
negative_counter = Counter()
neutral_counter = Counter()

for index, row in data.iterrows():
    opinions = row['opinions']
    for aspect, opinion in opinions.items():
        if opinion == 'positive':
            positive_counter[aspect] += 1
        elif opinion == 'negative':
            negative_counter[aspect] += 1
        else:
            neutral_counter[aspect] += 1

total_positive = sum(positive_counter.values())
total_negative = sum(negative_counter.values())
total_neutral = sum(neutral_counter.values())

if total_positive > total_negative and total_positive > total_neutral:
    final_opinion = 'Overall sentiment is positive.'
elif total_negative > total_positive and total_negative > total_neutral:
    final_opinion = 'Overall sentiment is negative.'
else:
    final_opinion = 'Overall sentiment is neutral.'

print(final_opinion)
```
